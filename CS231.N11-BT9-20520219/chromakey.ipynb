```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.imshow(img)
```

```python
background = img[:600,:750,:]
```

```python
plt.imshow(background)
```

```python
min_val = np.min(background,axis=(0,1))
```

```python
max_val = np.max(background,axis=(0,1))
```

```python
mask = cv.inRange(img,min_val,max_val)
```

```python
plt.imshow(mask)
```

```python
res = cv.bitwise_and(img,img,mask=mask)
```

```python
f = img - res
```

```python
plt.imshow(f)
```
